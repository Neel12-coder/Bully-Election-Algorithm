```python
def get_coordinator(process):
  max_process = -9999
  for process,status in process.items():
    if status == "SAFE":
      if process > max_process:
        max_process = process
  return max_process

def crash_process(process,process_id):
  process[process_id] = "CRASHED"

def get_crashed_processes(process):
  crashed_processes = []
  for process_id,status in process.items():
    if status == "CRASHED":
      crashed_processes.append(process_id)
  return crashed_processes

def recover_process(process,process_id):
  process[process_id] = "SAFE"

n = int(input("Enter the number of processes"))
process = dict()
for i in range(1,n+1):
  process[i] = "SAFE"
print("____________________________________________________")
print('Process STATUS')
for process_number, process_status in process.items():
    print('{}\t{}'.format(process_number, process_status))
print("____________________________________________________")
print()
print()
print()
option = 0 
while option!=4:
  print("*************************************************")
  print("1. Display Coordinator")
  print("2. Crash a Process")
  print("3. Recover a Crashed Process")
  print("4. Exit")
  print("*************************************************")
  option = int(input("Select an option"))
  if option == 1:
    print("The coordinator is {}".format(get_coordinator(process)))
  elif option == 2:
    p = int(input("Enter a process to crash"))
    if p == get_coordinator(process):
      print("The coordinator has crashed")
      print()
      initiator = int(input("Enter the initiator"))
      for process_id in process.keys():
        print("Process {} called for election".format(process_id))
      print()
      print()
      crash_process(process,p)
      for process_id,status in process.items():
        if status == "SAFE":
          print("Process {} is IN".format(process_id))
        else:
          print("Process {} is DEAD".format(process_id))
      print()
      print("**** The new coordinator is *****")
      print(get_coordinator(process))
    else:
      crash_process(process,p)
      print("The process {} has been crashed".format(p))
  elif option == 3:
    cp = get_crashed_processes(process)
    print("The crashed processes are ")
    print(*cp)
    rec_id = int(input("Enter the process to be recovered"))
    process[rec_id] = "SAFE"
    print("The process {} has been recovered".format(rec_id))
    print("The coordinator is {}".format(get_coordinator(process)))
  elif option == 4:
    print("Exited")
```
